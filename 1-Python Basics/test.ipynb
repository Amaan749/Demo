```python
##prime number
for num in range (1,101):
    if num>1:
        for i in range (2,num):
            if num%i==0:
                break
        else:
            print(num)
```

```python
##positive or negative no. 

num= float(input("enter your no."))
if num>0:
    print ("num is positive")
elif num==0:
    print ("num is zero")
else:
    print ("num is negative")
```
